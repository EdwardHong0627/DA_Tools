```python
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)
test_set.info()
```

```python
housing["income_cat"] = pd.cut(housing["median_income"], bins=[0,1.5,3,4.5,6,np.inf],labels=[1,2,3,4,5])
housing["income_cat"].hist()
```

```python
splits = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
housing.reset_index()
housing.head()
for train_index, test_index in splits.split(housing,housing["income_cat"]):
    strat_train_set=housing.loc[train_index]
    strat_test_set=housing.loc[test_index]
    
```

```python
strat_train_set.plot(kind="scatter", 
                     x="longitude", 
                     y="latitude", 
                     alpha=0.4, 
                     s=strat_train_set["population"]/100, 
                     c="median_house_value",
                     cmap=plt.get_cmap("jet"), 
                     colorbar=True,)
plt.legend()
```
